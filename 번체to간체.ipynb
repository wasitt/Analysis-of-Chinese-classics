```python
import pandas as pd
from zhconv import convert

# # 데이터 불러오기 및 컬럼명 설정
# df_huang = pd.read_excel("황정견DB_번체.xlsx", header=None).iloc[:, 0]
# df_huang.name = '황정견 시구'

# df_dubo = pd.read_excel("두보DB_번체.xlsx", header=None).iloc[:, 0]
# df_dubo.name = '두보 시구'
```

```python
df_huang=pd.read_excel("두황랜덤100수.xlsx",header=None,sheet_name="황랜덤").iloc[:,2]
```

```python
df_dubo=pd.read_excel("두황랜덤100수.xlsx",header=None,sheet_name="두랜덤").iloc[:,2]
```

```python
import re
```

```python
for i in range(len(df_huang)):
    if isinstance(df_huang.iloc[i], str):  # 문자열인지 확인
        line="".join(re.findall(r'[\u4E00-\u9FFF]+',df_huang.iloc[i]))
        df_huang.iloc[i] = convert(line, 'zh-hans')

for i in range(len(df_dubo)):
    if isinstance(df_dubo.iloc[i], str):  # 문자열인지 확인
        line="".join(re.findall(r'[\u4E00-\u9FFF]+',df_dubo.iloc[i]))
        df_dubo.iloc[i] = convert(line, 'zh-hans')
```

```python
df_huang.to_excel("황랜덤_간체.xlsx")
df_dubo.to_excel("두랜덤_간체.xlsx")
```
